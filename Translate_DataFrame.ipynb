```python
import pandas
from googletrans import Translator
tradutor = Translator()
```

```python
lista = ['Year','Milestone Title','Milestone Event','Why Important','Reference']
listacols = ['ano','descricao','eventos','observacao','link_referencia']

df = pandas.read_csv("data_analysis.csv", sep=",", usecols = lista)
df.columns = listacols
df.head(15)
```

```python
'''
essa função detecta se os dados da coluna são "en" de ingles 
e traduz para portugues
'''
def traduzir(value):
    if str(tradutor.detect(value))[14:16] == "en":
        td = tradutor.translate(value, src="en", dest="pt")
        return td.text
```

```python
'agora vamos traduzir os dados da coluna "Importancia"'
df['traduzido'] = df.observacao.map(traduzir)

'dados da coluna "observacao" traduzidos para "traduzido"'
df[['ano','descricao','traduzido']]
```
